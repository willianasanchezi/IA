```python
"""
It is a hybrid face recognition framework wrapping state-of-the-art models:

    Es un framework híbrido que combina y utiliza múltiples modelos de reconocimiento facial avanzados 
    (los más actuales y efectivos, llamados "state-of-the-art").

"""
print("")
```

```python
"""
VGG-Face, FaceNet, OpenFace, DeepFace, DeepID, ArcFace, Dlib, SFace and GhostFaceNet:

    Estos son los modelos avanzados y especializados que DeepFace usa o "envuelve". Cada uno de estos modelos ha sido desarrollado con diferentes enfoques y técnicas para realizar reconocimiento facial de manera muy precisa.
        VGG-Face: Un modelo desarrollado por la Universidad de Oxford para el reconocimiento facial.
        FaceNet: Un modelo de Google diseñado para transformar las caras en representaciones numéricas.
        OpenFace: Una herramienta de código abierto desarrollada por CMU para el reconocimiento facial.
        DeepFace: Un modelo de Facebook que popularizó el uso del deep learning para esta tarea.
        DeepID: Un modelo que utiliza redes neuronales profundas para el reconocimiento facial.
        ArcFace: Un modelo que mejora el reconocimiento facial mediante un enfoque geométrico avanzado.
        Dlib: Un conjunto de herramientas para el aprendizaje automático y procesamiento de imágenes.
        SFace y GhostFaceNet: Otros modelos que también han demostrado alta precisión en el reconocimiento facial.
"""
print("")
```
